```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file and convert to numbers
with open('save_coords.csv', mode='r') as file:
    reader = csv.reader(file)
    save_coords = [[float(value) for value in row] for row in reader]

# Calculate the ratio between the n and n+1 segment for each element of save_coords
order = [2,3,4,5,6,7,8]
for i in range(len(order)):

    ratio =  [save_coords[i][n+1] / save_coords[i][n] for n in range(1, len(save_coords[i]) - 1)]

    print(f'Mesh with 2**{order[i]}: {save_coords[i]}')
    print(f'Ratios = {ratio}')
    print()

    plt.figure()
    plt.plot(range(1, len(save_coords[i])-1), ratio, 'o-')
    plt.xlabel('Element')
    plt.ylabel('Ratio between n+1 and n segment')
```
